```python
import numpy as np

# Task 2: Array Indexing and Slicing

# 1. Create a NumPy array from the following list:

data = [10, 20, 30, 40, 50, 60, 70, 80, 90]
arr2 = np.array(data)

# 2. Retrieve the first three elements of the array.

first_three = arr2[:3]
print("First three elements:", first_three)

# Retrieve every alternate element of the array.

alternate_elements = arr2[::2]
print("Every alternate element:", alternate_elements)

# Reverse the array.
reverse = arr2[::-1]
print("Reversed array:", reverse)
```

```python
import numpy as np

# Task 3: Mathematical Operations

# 1. Create two NumPy arrays, A and B, each with 5 random integers between 1 and 20.

A = np.random.randint(1, 21, 5)
B = np.random.randint(1, 21, 5)

print("Array A:", A)
print("Array B:", B)

# 2. Add, subtract, multiply, and divide the two arrays element-wise.

print("\nA + B:", A + B)
print("A - B:", A - B)
print("A * B:", A * B)
print("A / B:", A / B)

# Compute the dot product of the arrays.

print("\nDot product:", np.dot(A, B))

# Find the mean, median, standard deviation, and variance of array A.

print("\nMean of A:", np.mean(A))
print("Median of A:", np.median(A))
print("Standard deviation of A:", np.std(A))
print("Variance of A:", np.var(A))

# Identify the maximum and minimum values in array B and their indices.

print("\nMax of B:", np.max(B), "at index", np.argmax(B))
print("Min of B:", np.min(B), "at index", np.argmin(B))
```

```python
import numpy as np

# Task 4: Reshaping and Transposing

# 1. Create a 1D NumPy array of 12 integers ranging from 1 to 12.
arr = np.arange(1, 13)
print("1D array: ", arr)

# 2. Reshape the array into a 2D array of shape (4, 3).

arr_2d = arr.reshape(4, 3)
print("\nReshaped to 2D (4x3):\n\n", arr_2d)

# Reshape the array into a 3D array of shape (2, 2, 3).

arr_3d = arr.reshape(2, 2, 3)
print("\nReshaped to 3D (2x2x3):\n\n", arr_3d)

# Transpose the reshaped 2D array and display its shape.
Transpose_2d = arr_2d.T
print("\nTransposed 2D array:\n\n",Transpose_2d )
print("\nShape of transposed 2D array:", Transpose_2d.shape)
```

```python
import numpy as np

# Task 5: Boolean Masking and Filtering

#1. Create a NumPy array with 15 random integers between 10 and 50.

array = np.random.randint(10,51,15)
print("Original Array:\n", array)

#2. Write a Python program to:

# Find all elements greater than 25.

elements = array[array>25]
print("\nElements greater than 25:\n", elements)

# Replace all elements less than 30 with 0.

array[array < 30] = 0
print("\nArray after replacing elements < 30 with 0:\n", array)

# Count the number of elements divisible by 5.

count = np.sum(array % 5 == 0)
print("\nNumber of elements divisible by 5:", count)
```

```python
import numpy as np
import time

#Task 7: Generic

# 1.Create a Python program that:


# Generate two large random arrays of size (100, 100)

st_time = time.time()
A = np.random.rand(100, 100)
B = np.random.rand(100, 100)



# Perform matrix multiplication

C = np.dot(A,B)

# Find the determinant

det = np.linalg.det(C)

# We can find the inverse if determinant is not zero

if det != 0:
	inv = np.linalg.inv(C)
	print("Inverse calculated successfully.\n")
else:
	inv = None
	print("Matrix is singular, inverse does not exist.")

end_time = time.time()

print("Array 1\n\n", A)
print("\nArray 2\n\n", B)
print("\nDeterminant:", det)
print("\nThe Inverse is", inv)
print("\nTime taken (seconds):", end_time - st_time)
```
