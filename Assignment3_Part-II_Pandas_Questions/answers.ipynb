```python
import pandas as pd
import numpy as np

# Task 1: Working with Series

# 1. Create a Pandas Series from the following list:

data = [25, 30, 35, 40, 45]
series = pd.Series(data)
print("Pandas Series:\n", series)

# 2. Assign custom indices ['A', 'B', 'C', 'D', 'E'] to the Series

updated_series = pd.Series(data, index=['A', 'B', 'C', 'D', 'E'])
print("\nUpdated Series with custom indices:\n", updated_series)

# Display the first three elements of the Series

print("\nFirst 3 elements:\n", updated_series[:3])

# Calculate mean, median, and standard deviation

mean_val = updated_series.mean()
median_val = updated_series.median()
std_dev = updated_series.std()


print("Mean:", mean_val)
print("Median:", median_val)
print("Standard Deviation:", std_dev)
```

```python
import pandas as pd

# Task 3: Data Selection and Filtering

# 1.Using the DataFrame from Task 2, write a Python program to:

data = {
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eve'],
    'Age': [20, 22, 19, 21, 20],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female'],
    'Marks': [85, 78, 92, 74, 88]
}

df = pd.DataFrame(data)

#Select and display the Name and Marks columns.

print("Name and Marks columns:\n")
print(df[['Name', 'Marks']])

#Filter and display records where Marks > 80.
print("\nStudents with Marks > 80:\n")
print(df[df['Marks'] > 80])

#Display the record of the student with the highest marks.

print("\nStudent with the Highest Marks:\n")
print(df[df['Marks'] == df['Marks'].max()])
```

```python
import pandas as pd

# Task 4: Handling Missing Data

# Re-create the original DataFrame from Task 2

data = {
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eve'],
    'Age': [20, 22, 19, 21, 20],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female'],
    'Marks': [85, 78, 92, 74, 88]
}

df = pd.DataFrame(data)

# 1. Introduce missing values

df.loc[1, 'Marks'] = None   
df.loc[4, 'Age'] = None     

print("DataFrame with Missing Values:\n")
print(df)

# 2. Identify missing values

print("\nMissing Values (True indicates missing):\n")
print(df.isnull())

# 3. Fill missing values in 'Marks' column with mean of the column

mean_marks = df['Marks'].mean()
df['Marks'] = df['Marks'].fillna(mean_marks)


print("\nDataFrame after filling missing Marks with mean:\n")
print(df)

# 4. Drop rows where 'Age' is missing

new_df = df.dropna(subset=['Age'])

print("\nFinal DataFrame after dropping rows with missing Age:\n")
print(new_df)
```

```python
import pandas as pd

# Task 5: Grouping and Aggregation

# Re-create the original DataFrame (from Task 2)

data = {
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eve'],
    'Age': [20, 22, 19, 21, 20],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female'],
    'Marks': [85, 78, 92, 74, 88]
}

df = pd.DataFrame(data)

# 1. Group by Gender and calculate mean Age and Marks

grouped_mean = df.groupby('Gender')[['Age', 'Marks']].mean()
print("Mean Age and Marks by Gender:\n")
print(grouped_mean)

# 2. Count number of students in each gender group

gender_counts = df['Gender'].value_counts()
print("\nNumber of Students by Gender:\n")
print(gender_counts)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/ShubhamSinghal12/GLA_pythonML2025/refs/heads/main/Data%20Visualization/Titanic.csv"
df = pd.read_csv(url)

# Summary Statistics

print("\nSummary Statistics:\n\n", df.describe(include='all'))

# Missing Value Check

print("\nMissing Values Count:\n\n", df.isnull().sum())



# Data Visualization: Plots and Curves


# Age Distribution

plt.figure(figsize=(7,4))
sns.histplot(df['Age'], bins=30, kde=True, color='skyblue')
plt.title("Age Distribution with Curve (KDE)")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()


# Fare Distribution

plt.figure(figsize=(7,4))
sns.histplot(df['Fare'], bins=30, kde=True, color='lightgreen')
plt.title("Fare Distribution with Curve (KDE)")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()


# Gender vs Survival

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Sex', hue='Survived', palette='Set2')
plt.title("Gender vs Survival")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()

# Pclass vs Survival

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Pclass', hue='Survived', palette='Set1')
plt.title("Class vs Survival")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()


# Embarked vs Survival

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Embarked', hue='Survived', palette='Set3')
plt.title("Embarked Port vs Survival")
plt.xlabel("Port")
plt.ylabel("Count")
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()


# Heatmap of Correlation

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()


# ---------------------------
# Findings
# --------------------------


# Missing Values

#  Age has 177 missing values — important for future modeling or imputation.
#  Cabin is mostly missing — may not be useful without feature engineering.
#  Embarked has only 2 missing entries — can be safely filled with the mode.



# 1. Age Distribution

#  Most passengers were between 20–30 years old.
#  Few were very old or very young.
#  The shape is right-skewed.

# 2. Fare Distribution

#  Many passengers paid low fares.
#  A few paid very high fares.
#  The fare is highly skewed.

# 3. Gender vs Survival

#  More females survived than males.
#  Males had higher death rates.

# 4. Pclass vs Survival

#  1st class passengers survived more.
#  3rd class had the least survival.

# 5. Embarked vs Survival

#  Most passengers got on at port 'S' (Southampton).
#  At port 'C' (Cherbourg), the number of survivors was higher than non-survivors,
#   which is not the case for other ports.
#  Port 'Q' (Queenstown) had the lowest number of passengers and survivors.

# 6. Correlation Heatmap (Key Points)

#  Fare is positively correlated with survival.
#  Pclass is negatively correlated with survival.
```
